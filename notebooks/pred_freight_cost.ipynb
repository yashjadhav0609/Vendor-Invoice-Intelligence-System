## Predicting Freight Costs with Machine Learning

In the logistics industry, accurately predicting freight costs is crucial for budgeting, pricing strategies, and overall operational efficiency. Machine learning offers powerful tools to analyze historical data and make informed predictions about future freight costs.

**Objective:** Predict freight cost for a vendro invoice using quantity and dollars, to improve budgeting and cost management.

- It's a non-trivial component of landed cost.
- Poor freight estimates distort margin and inventory management.
- Automating freight estimation helps procurement teams forecast true cost before invoice arrival.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 42)
```

```python
X_train
```

```python
# Test data on three models (Linear Regression, Decision Tree Regressor, Random Forest Regressor) then evaluate.

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score
```

```python
model1 = LinearRegression()
model1.fit(X_train, y_train)

model2 = DecisionTreeRegressor(max_depth = 5, random_state = 42)
model2.fit(X_train, y_train)

model3 = RandomForestRegressor(max_depth = 3, random_state = 42)
model3.fit(X_train, y_train)
```

```python
# Evaluate Model to cheak MAE, RMSE, R2_score.

def evaluate_model(model, X_test, y_test, model_name):
    preds = model.predict(X_test)

    mae = mean_absolute_error(y_test, preds)
    rmse = root_mean_squared_error(y_test, preds)
    r2 = r2_score(y_test, preds)*100

    print(f"\n{model_name} Performance:")
    print(f"MAE     : {mae:.2f}")
    print(f"RMSE    : {rmse:.2f}")
    print(f"R2      : {r2:.2f}%")
```

```python
evaluate_model(model1, X_test, y_test, 'Linear Regression')
evaluate_model(model2, X_test, y_test, 'Decision Tree Regression')
evaluate_model(model3, X_test, y_test, 'Random Forest Regression')
```
